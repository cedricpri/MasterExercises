#### 26. What is the output of the following script? (★☆☆)

```python
# Author: Jake VanderPlas

print(sum(range(5),-1))
from numpy import *
print(sum(range(5),-1))
```

````python
print(sum(range(5),-1))
from numpy import *
print(sum(range(5),-1))

# why?
# Apparently, the two lines do not give back the same value after importing * from numpy (at least the first we are executing the cell)
# When we do from numpy import *, all the numpy functions will be accessible into the local namespace, which means that the usual sum() function is overwritten
# When using the "usual" sum function, the sum will start at -1, the second argument given, given us 9 instead of 10
# Is it usually better to do from numpy import np to be able to access both sum functions: sum() and np.sum()
````

#### 29. How to round away from zero a float array ? (★☆☆)

````python
# np.copysign
# Copysign(x1, x2) changes the sign of x1 to that of x2, element-wise.
a = np.array([-1.5, -1.0, -0.5, 0-0, 0.5, 1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0])
print(np.copysign(np.ceil(np.abs(a)), a))
````

#### 32. Is the following expressions true? (★☆☆)

```python
np.sqrt(-1) == np.emath.sqrt(-1)
```

````python
np.sqrt(-1) == np.emath.sqrt(-1)
# why?
# It does not give the same value
# This is because using the emath domain allows us to consider complex values (while we get a nan on the left hand side)
# It is easy to check by checking the value on each side of the expression:
print(np.sqrt(-1))
print(np.emath.sqrt(-1))
# We can clearly see that the two expressions are not equal
````

## SQL IO

Create a database called 'billionaries' using sqlite and save the data from the dataframe using the appropiate tables. Save the information of people, companies and the relationship between people and companies (ie. company.relationship). Use the functions viewed in class.

````python
from sqlalchemy import create_engine
from sqlalchemy import MetaData, Table, Column, Integer, String, Float, Boolean, ForeignKey

engine = create_engine('sqlite:///billionaries.db')

#Now let's create the data frames we will need
df_people = df[['name', 'demographics.gender']]
df_people = df_people.drop_duplicates(keep = 'first') #Drop the duplicates
df_people = df_people.reset_index()
df_people.columns = ['df_number', 'name', 'gender']
df_people['id_people'] = df_people.index
print(df_people.head())

df_companies = df[['company.name', 'company.founded', 'company.sector', 'company.type', 'company.relationship']]
df_companies = df_companies.drop_duplicates(keep = 'first') #Drop the duplicates
df_companies = df_companies.reset_index()
df_companies.columns = ['id_people', 'name', 'founded', 'sector', 'type', 'relationship']
df_companies['company_id'] = df_companies.index
print(df_companies.head())
````

````python
#Let's now define the positions data frame by merging the other two data frames
df_positions = pd.merge(df_people, df_companies, left_on=df_people['df_number'], right_on=df_companies['id_people'])
df_positions = df_positions[['df_number','company_id','relationship']]
df_companies_new = df_companies.drop(columns = ['relationship'])
print(df_positions.head())
````

````python
#Now let's convert these data frames to sql
df_people.to_sql('People', con=engine, if_exists='replace')
df_companies_new.to_sql('Companies', con=engine, if_exists='replace')
df_positions.to_sql('Positions', con=engine, if_exists='replace')
````

Extract necessary information:

````python
print(engine.execute("SELECT * FROM People").fetchall())
print(engine.execute("SELECT * FROM Companies").fetchall())
print(engine.execute("SELECT * FROM Positions").fetchall())
````
